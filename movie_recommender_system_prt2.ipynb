```python
import pandas as pd
import numpy as np
import pickle
```

```python
new_df=pd.read_csv('/kaggle/input/new-dfinfinity-last/new_df.csv')
df5=pd.read_csv('/kaggle/input/5000mvi/tmdb_5000_movies.csv')
```

```python
new_df=new_df[~(new_df['title'].str.match(r'^[^a-zA-Z0-9]'))]  #removing these movies cuse no one gonna searching these
```

```python
new_df.shape
df5.shape
```

```python
k=df5[df5['original_language']=='en']
```

```python
k.shape
```

```python
except_new_df=new_df[~(new_df['id'].isin(k['id']))]       
except_new_df.reset_index(inplace=True)
except_new_df=except_new_df.iloc[:8104]
except_new_df
```

```python
new_df=new_df[(new_df['id'].isin(k['id']))|(new_df['id'].isin(except_new_df['id']))]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features= 5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)  # caution if you are runnng this code then it might take a lotof space in your ram and time as well .
```

```python
similarity.shape
```

```python
pickle.dump(similarity,open('similarity.pkl','wb'))  #you need this file to use in app.py file (file-1)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie] .index[0]
    distances = similarity[movie_index]
    movieslist = sorted(list(enumerate(distances)),reverse=True, key= lambda x:x[1])[1:6]
    
    for i in movieslist:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
